### Special Instructions: 
undo          - undo the last natural language instruction. <br>
redo          - redo the last undo. <br>
reset         - reset the plot specs, data and undo/redo stacks. <br>
plot          - draw the graph with current plot specs and data. <br>
exit          - end the current session with the plotting agent. <br>
next          - after loading source dialog, switch to the next dialog in the dataset. <br>
generate data - generate a dataset from data sampler according to the current plot specs. <br>
set labels    - set the labels of the plot manually. <br>
load source dialog - load dialogs from the training/test set whose path is specified by src_address. <br>
load csv - load data from a csv file. (user's upload is not supported in Binder. Enter "time_series_covid19_confirmed_global_simplified.csv" for the path) <br>
load dialog - load customized dialogs from a file. (user's upload is not supported in Binder. Enter "dialog_test" for the path) <br>
